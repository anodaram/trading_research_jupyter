```python
import pyfolio as pf
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline


# silence warnings
import warnings
warnings.filterwarnings('ignore')
```

### Load data from catalyst\zipline

```python
perf = pd.read_csv('perf_TradingPair(452516 [eth_btc]).csv', index_col = 'Unnamed: 0')
perf.index = pd.to_datetime(perf.index)
perf.head()
```

### Look at stats

```python
pf.show_perf_stats(perf.returns)
```

```python
pf.show_worst_drawdown_periods(perf.returns)
```

### Let's draw some graphs

```python
plt.figure(figsize = (15, 10))
plt.plot(perf.benchmark_period_return)
plt.plot(perf.algorithm_period_return)
plt.legend()
plt.ylabel('Equity')
plt.xlabel('Date')
```

```python
plt.figure(figsize = (15, 10))
pf.plot_drawdown_underwater(perf.returns).set_xlabel('Date')
```

```python
plt.figure(figsize = (15, 10))
pf.plot_drawdown_periods(perf.returns, top = 5).set_xlabel('Date')
```

```python
plt.figure(figsize = (15, 10))
pf.plot_returns(perf.returns).set_xlabel('Date')
```

```python
plt.figure(figsize = (15, 10))
pf.plot_return_quantiles(perf.returns).set_xlabel('Timeframe')
```

```python
plt.figure(figsize = (15, 10))
pf.plot_monthly_returns_dist(perf.returns).set_xlabel('Returns')
```

```python
plt.figure(figsize = (15, 10))
pf.plot_rolling_volatility(perf.returns, rolling_window = 30).set_xlabel('Date')
```
